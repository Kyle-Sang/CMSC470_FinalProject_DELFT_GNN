```python
corpus = []
titles = []
for a in tqdm(range(ord('A'),ord('F') + 1)):
    for b in tqdm(range(ord('A'),ord('Z') + 1)):
        for x in range(00, 99 + 1):
            if chr(a) == 'F' and chr(b) == 'Z' and x > 16:
                break
            filename = chr(a) + chr(b) + '/wiki_' + ('%02d' % x) + '.bz2'
            with bz2.open("wikiAbstracts/" + filename, "r") as f:
                content = f.read().decode().split('\n')
            for page in content[:-1]:
                jPage = json.loads(page)
                corpus.append("".join(jPage['text']))
                titles.append(jPage['title'])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(corpus)
```

```python
question = 'President penny monument top hat.'
```

```python
question_tfidf = vectorizer.transform([question])
cosine_similarities = cosine_similarity(question_tfidf, X)
cos = cosine_similarities[0]
indices = cos.argsort()[::-1]
guesses = []
for i in range(5):
    idx = indices[i]
    guess =  {"title": titles[idx], "confidence": cos[idx]}
    guesses.append(guess)
```

```python
class wikiReader():
    def __init__(self, info):
        self.info = info
    
    def __str__(self) -> str:
        return str(self.info[0]) + '.' + ('%02d' % self.info[1]) + '.' + ('%02d' % self.info[2]) + '.' + ('%03d' % self.info[3])
    
    def read(self) -> bytes:
        with open("wikiAbstracts/" + str(self.info[0]) + "/wiki_" + ('%02d' % self.info[1]), "r") as f:
            pages = f.read().split('\n')[:-1]
            jPage = json.loads(pages[self.info[2]])
        return "".join(jPage['text'])
    
    def getOg(self) -> json:
        with open("wikiAbstracts/" + str(self.info[0]) + "/wiki_" + ('%02d' % self.info[1]), "r") as f:
            pages = f.read().split('\n')[:-1]
            return json.loads(pages[self.info[2]])
```

```python
class wikiReaderStr():
    def __init__(self, info):
        self.info = info
    
    def __str__(self) -> str:
        return self.info
    
    def read(self) -> bytes:
        with bz2.open("wikiAbstracts/" + str(self.info[0:2]) + "/wiki_" + self.info[2:4] + ".bz2", "r") as f:
            content = f.read()
            s = content.decode()
            pages = s.split('\n')[:-1]
            jPage = json.loads(pages[int(self.info[4:6])])
        return "".join(jPage['text'])
```

```python
corpusNC = []
for a in tqdm(range(ord('A'),ord('F') + 1)):
    for b in tqdm(range(ord('A'),ord('Z') + 1)):
        for x in range(00, 99 + 1):
            if chr(a) == 'F' and chr(b) == 'Z' and x > 16:
                break
            filename = chr(a) + chr(b) + '/wiki_' + ('%02d' % x)
            with open("wikiAbstracts/" + filename, "r") as f:
                content = f.read().split('\n')[:-1]
            for pageii in range(0,len(content)):
                corpusNC.append(wikiReader((chr(a) + chr(b), x, pageii, 0)))
```
